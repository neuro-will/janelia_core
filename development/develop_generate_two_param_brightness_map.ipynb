```python
import numpy as np
import matplotlib.pyplot as plt

from janelia_core.visualization.custom_color_maps import generate_two_param_hsv_map
from janelia_core.visualization.custom_color_maps import make_red_blue_green_c_map
```

## Parameters go here

```python
clr_param_range = (2, 3, .001)
vl_param_range = (4, 6, .005)
```

```python
clrs = generate_two_param_hsv_map(clr_param_range=clr_param_range, vl_param_range=vl_param_range, 
                                  p1_cmap=make_red_blue_green_c_map(1024), 
                                  clims=(clr_param_range[0], clr_param_range[1] - clr_param_range[2]),
                                  vllims=(vl_param_range[0], vl_param_range[1] - vl_param_range[2]))
```

## View the resulting color map

```python
n_clr_smps = len(np.arange(*clr_param_range))
n_vl_smps = len(np.arange(*vl_param_range))

clr_smp_params = np.repeat(np.expand_dims(np.arange(*clr_param_range), 1), n_vl_smps, axis=1)
vl_smp_params = np.repeat(np.expand_dims(np.arange(*vl_param_range), 0), n_clr_smps, axis=0)
```

```python
smp_clrs = clrs[clr_smp_params, vl_smp_params]
```

```python
plt.imshow(smp_clrs, extent=[vl_param_range[0], vl_param_range[1],
                             clr_param_range[1],  clr_param_range[0]])
```
