```python
import matplotlib.pyplot as plt
import numpy as np

from janelia_core.stats.regression import naive_regression
```

```python
%matplotlib notebook
```

## Parameters go here

```python
x_dim = 10
y_dim = 2
n_smps = 10000
noise_std = 1
```

## Generate data

```python
# Standard deviation of different x variables
x_std = 10*np.random.rand(10) + 1
x_mns = 5*np.random.randn(10)
x = np.random.randn(n_smps, x_dim)*x_std + x_mns

beta = np.random.randn(x_dim, y_dim)

y_true = np.matmul(x, beta)
y = y_true + np.random.randn(n_smps, y_dim)*noise_std
```

## Fit the model

```python
beta, o = naive_regression(x=x, y=y)
```

## Get prediction error

```python
y_hat = np.matmul(x, beta) + o
er = y_hat - y
mse = np.mean(er**2)
print('MSE: ' + str(mse))
```

## Examine predictions

```python
plot_var = 0
plt.plot(y_true[:, plot_var], 'ro')
plt.plot(y_hat[:, plot_var], 'b.')
```
