```python
import matplotlib.pyplot as plt
import numpy as np

from janelia_core.math.basic_functions import pts_in_arc
```

```python
arc_angle = [-.5*np.pi - 8*np.pi, -.25*np.pi- 8*np.pi]
ctr = np.asarray([.5,.5])
pts = np.random.rand(1000,2)
```

```python
arc_pts = pts_in_arc(pts=pts, ctr=ctr, arc_angle=arc_angle)
```

```python
plt.figure()
ax = plt.subplot(1,1,1)
ax.plot(pts[arc_pts,0], pts[arc_pts,1],  'r.')
ax.plot(pts[~arc_pts,0], pts[~arc_pts,1],  'k.')
ax.axis('equal')
```
