# Script to exlore different ways of viewing a scatter plot

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
n_smps = 10000
smps = np.random.normal(size=(3, n_smps))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(smps[0,], smps[1,], smps[2,], marker=",")
plt.show()
```
